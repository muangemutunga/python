# 🌸 Analyzing Data with Pandas and Visualizing Results with Matplotlib

## 📘 Objective
- Load and analyze the Iris dataset using pandas.
- Create basic plots and visualizations using matplotlib and seaborn.
- Discover insights through data exploration and visual representation.

```python
# 📚 Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Enable inline plotting
%matplotlib inline
sns.set(style="whitegrid")
```

## 📥 Task 1: Load and Explore the Dataset

```python
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display the first few rows
df.head()
```

```python
# Check data structure and missing values
print(df.info())
df.isnull().sum()
```

✅ The dataset is clean and has no missing values.

## 📊 Task 2: Basic Data Analysis

```python
# Basic statistics
df.describe()
```

```python
# Group by species and calculate means
df.groupby("species", observed=True).mean()
```

### 🔍 Observations:
- *Setosa* has the smallest petal and sepal sizes.
- *Virginica* generally has the largest measurements.
- *Versicolor* falls in between.

## 📈 Task 3: Data Visualization

```python
# 1. Line Chart (simulated time trend)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length", color='green')
plt.title("Simulated Sepal Length Over Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart (Average Petal Length per Species) — Future-proof version
plt.figure(figsize=(8, 5))
sns.barplot(
    x='species',
    y='petal length (cm)',
    hue='species',  # Explicitly set hue
    data=df,
    estimator=np.mean,
    palette="viridis",
    legend=False    # Avoid duplicate legends
)
plt.title("Average Petal Length per Species")
plt.ylabel("Average Petal Length (cm)")
plt.xlabel("Species")
plt.show()
```

```python
# 3. Histogram (Sepal Width Distribution)
plt.figure(figsize=(8, 5))
plt.hist(df["sepal width (cm)"], bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Scatter Plot (Sepal Length vs Petal Length)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species", palette="Set1")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()
```

## ✅ Summary of Insights
- Iris dataset shows clear distinction between species based on petal size.
- *Setosa* is very distinct, with smallest petals.
- Scatter plots show good potential for classification based on these features.
